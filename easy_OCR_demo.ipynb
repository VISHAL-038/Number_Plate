```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
```

```python
Image("/content/fancy_number_plate_bfbc501f34 copy.jpg")
```

```python
render = easyocr.Reader(['en'])
```

```python
output = render.readtext('/content/fancy_number_plate_bfbc501f34 copy.jpg')
```

```python
output
```

```python
cord = output[-1][0]
```

```python
cord
```

```python
a = list(zip(*cord))
```

```python
a
```

```python
min(a[0])
```

```python
min(a[1])
```

```python
x_min,y_min = [int(min(idx)) for idx in zip(*cord)]
```

```python
x_min , y_min
```

```python
x_max,y_max = [int(max(idx)) for idx in zip(*cord)]
```

```python
x_max,y_max
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,30
```

```python
image = cv2.imread('/content/fancy_number_plate_bfbc501f34 copy.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
